####List Functions

Below are some of the common list functions in python. There are several others whose details can be found at https://www.programiz.com/python-programming/methods/list

```
#List Functions

#1. length of list
print("1: ",len(cash_flows_1_Bond))

#2. last element of list
print("2: ",cash_flows_1_Bond[-1])

#3. Append: to insert at the end
total_cashflow_portfolio = []
total_cashflow_portfolio.append(30)
print("3: ",total_cashflow_portfolio)
total_cashflow_portfolio.append(20)
print("4: ",total_cashflow_portfolio)
total_cashflow_portfolio.append(-10)
print("5: ",total_cashflow_portfolio)
total_cashflow_portfolio.append(40)
print("6: ",total_cashflow_portfolio)

#4. Insert: To insert at a specific location
total_cashflow_portfolio.insert(1,35)
print("7: ",total_cashflow_portfolio)

#5. Pop
total_cashflow_portfolio.pop(2) #pops the specific element
print("8: ",total_cashflow_portfolio)
total_cashflow_portfolio.pop() #pops the last element
print("9: ",total_cashflow_portfolio)

#6: Copy: 
##If you assign a list to a new variable and modify the new variable, the old list also gets modified
new_list = total_cashflow_portfolio
print("10: ", new_list)
new_list.append(70)
print("11: ", new_list)
print("12: ", total_cashflow_portfolio)

##To avoid this issue and retain the original list while modifying the new list, use list.copy
new_list_copy = list.copy(total_cashflow_portfolio)
print("13: ", new_list_copy)
new_list_copy.append(100)
print("14: ", new_list_copy)
print("15: ", total_cashflow_portfolio)
```

Operations on custom list

```
# Print square of each residual of OLS regression
residuals = [0.2, 0.4, -0.5, 0.3]
for i in residuals:
  print(i*i)
```

```
# Print square of each residual assigning the index to the result and also print sum of square residuals of OLS regression
residuals = [0.2, 0.4, -0.5, 0.3]
sum = 0
cnt = 1

for i in residuals:
  print("Square Residual ", cnt, " = ", i*i)
  sum = sum + i*i
  cnt = cnt+1
print("Sum of square residual is", sum)
```

```
# Create a list of squared residuals
square_residuals = []
residuals = [0.2, 0.4, -0.5, 0.3]

for i in residuals:
  square_residuals.append(i*i)

print(square_residuals) #make sure that the indentation is correct. If this is indented within the for loop, then the results will be printed four times
```

###2. While Loop

```
#While loops may be used in optimization problems (with tolerance threshold in bisection)
initial_range = 40
tolerance = 0.02 

while (initial_range>tolerance):
  print("New Value = ", initial_range)
  initial_range = initial_range/2

print(initial_range)
```

##**User defined Functions**

```
#create a function for sum of first n numbers (user input is n):

def sum_first_num(n):
  sum = 0 #without initializing, it will through en error (UnboundLocalError: local variable 'sum' referenced before assignment)
  for i in range(n):
    sum = sum+ (i+1)
  return sum
```

```
a = sum_first_num(20)
b = a*a
print(b)
```

```
#create a function for factorial

def fact_val(n):
  fact = 1
  for i in range(n):
    fact = fact*(i+1)
  return fact
```

```
second_order_coeff = 1/ fact_val(2)
second_order_coeff
```
