# The Data Matrix and the Principal Component Analysis

* The Principal Component Analysis is a popular statistical technique

    * Transform a set of data X into a new orthogonal basis 
    * Directions of the new basis are ordered in terms of variance contribution to the overall variance.

* We consider a $p$-dimensional vector of features observed $n$ times 

* Let us denote the $n\times p$ dimensional matrix of data as $X$.

***

***

# The Data Matrix

* In general, the data matrix $X$ is rectangular 

* Eigenvalues are not defined for rectangular matrices. 

* Singular values: 

    * The right and left singular values for a rectangular matrix $X$ are the square roots of the eigenvalues of $X^{T}X$ and $XX^{T}$. 

* The corresponding eigenvectors – the right and left singular vectors, respectively. 

What exactly do these matrices, $M^{T}M$ and $MM^{T}$ mean in terms of statistics?

* $x$ corresponds to observations – $p$-dimensional vector random variable

* We assume $x$ is a column vector $x\in\Re^{p\times1}$

* The samples---realisations of the random variable $x$---occur as the rows of $X$, not as columns
***

***
# PCA: The First Principal Component


* Let us introduce the Principal Components of the data. 

* We focus on the calculation of the first principal component.

* We consider dataset with $p$ features. 

     * We model our data as a $p$-dimensional random variable ${x}\in\Re^{p\times1}$.

* The objective of the first principal component:

    * To find a vector of weights ${w}_{(1)}\in\Re^{p\times1}$ such that the variance of the scalar random variable ${x^{T}}{w}_{(1)}$ is maximized.

    * Thus our objective is to find 

$$ \max\limits _{{w}_{(1)}}\,Var[x^{T}{w}_{(1)}]\,.$$

***

***
# PCA: The First Principal Component


* We assume that we are given a data matrix with n observations, ${X}\in\Re^{n\times p}$. 

* Let us consider the centred data matrix ${X}'$. 

* We have already argued that the sample covariance matrix: 

$${Q}=\frac{1}{n-1}\left({X}'\right)^{T}{X}'$$

is an unbiased estimator of the covariance matrix $Var[{x}]$. 

* Moreover, by the properties of variances, 

$$Var[x^{T}{w}_{(1)}]={w^{T}}_{(1)}Var[{x}]{w}_{(1)}={w^{T}}_{(1)}{\Sigma}{w}_{(1)}\,.$$

* In practice, however, we don't have access to ${\Sigma}$ – at best, we can get unbiased estimator ${C}$, namely the sample covariance matrix: 

$${C}={Q}=\frac{1}{n-1}\left({X}'\right)^{T}{X}'\,.$$

* Thus our maximization problem becomes: 

$$\max\limits _{{w}_{(1)}}\,{w^{T}}_{(1)}{C}{w}_{(1)}\,.$$

***

***
# PCA: The First Principal Component

* Let us add another requirement (another constraint): we want our weights, ${w}_{(1)}$, to be a unit vector: 

$${w^{T}}_{(1)}{w}_{(1)}=1\,.$$

* Equivalently, we can add a penalty to the objective function for any deviation from zero of ${w^{T}}_{(1)}{w}_{(1)}-1$: 

$$\max\limits _{{w}_{(1)}}\,{w^{T}}_{(1)}{C}{w}_{(1)}-\lambda_{(1)}({w^{T}}_{(1)}{w}_{(1)}-1)\,,$$

where the constant $\lambda_{(1)}$ is a Lagrange multiplier.

* Let us now solve this optimization problem:

$$\frac{\partial}{\partial{w}_{(1)}}\left({w^{T}}_{(1)}{C}{w}_{(1)}-\lambda_{(1)}({w^{T}}_{(1)}{w}_{(1)}-1)\right)=0\,,$$

* computing the partial derivative on the left-hand side and rearranging this equation:

$${C}{w}_{(1)}=\lambda_{(1)}{w}_{(1)}\,.$$



***


***
# PCA: The Second Principal Component


* Objective: We look for a vector of weights ${w}_{(2)}\in\Re^{p\times1}$ such ${w}_{(2)}$ is orthogonal to ${w}_{(1)}$ and the variance of the scalar random variable $x^{T}{w}_{(2)}$ is maximized. 

    * We want this vector to be a unit vector, ${w^{T}}_{(2)}{w}_{(2)}=1$. 

    * Since it is orthogonal to ${w}_{(1)}$, ${w^{T}}_{(1)}{w}_{(2)}=0$. 

    * We can add this as a constraint to our optimization problem: 

$$\max\limits _{{w}_{(2)}}\,{w^{T}}_{(2)}{C}{w}_{(2)}-\lambda_{(2)}({w^{T}}_{(2)}{w}_{(2)}-1)-\nu_{(1)}{w^{T}}_{(1)}{w}_{(2)}\,,$$

* where $\lambda_{(2)}$ is the Lagrange multiplier corresponding to the unit vector constraint and $\nu_{(1)}$ the Lagrange multiplier corresponding to the orthogonality constraint.



***




***
# PCA: The (any) Principal Component

We can generalize this process and obtain:

* The eigenvalue, $\lambda_{(1)}$, and the corresponding eigenvector ${w}_{(1)}$, of ${C}$, such that $Var[{x^{T}}{w}_{(1)}]\approx{{w^{T}}_{(1)}}{C}{w}_{(1)}=\lambda_{(1)}$; 

* The eigenvalue, $\lambda_{(2)}\leq\lambda_{(1)}$, and the corresponding eigenvector ${w}_{(2)}$, of ${C}$, such that $Var[{x^{T}}{w}_{(2)}]\approx{{w^{T}}_{(2)}}{C}{w}_{(2)}=\lambda_{(2)}$; 

* The eigenvalue, $\lambda_{(3)}\leq\lambda_{(2)}$, and the corresponding eigenvector ${w}_{(3)}$, of ${C}$, such that $Var[{x^{T}}{w}_{(3)}]\approx{{w^{T}}_{(3)}}{C}{w}_{(3)}=\lambda_{(3)}$; 

* $\vdots$ 

* The eigenvalue, $\lambda_{(p)}\leq\lambda_{(p-1)}$, and the corresponding eigenvector ${w}_{(p)}$, of ${C}$, such that $Var[{x^{T}}{w}_{(p)}]\approx{{w^{T}}_{(p)}}{C}{w}_{(p)}=\lambda_{(p)} $

# PCA-based Dimensionality Reduction, Clustering and Outlier Detection

## Dimensionality Reduction

* The $p$-dimensional dataset

    * Calculate first $k$ Principal Components (first $k$ largest eigenvalues and corresponding eigenvectors)
    * Obtain first $k$ projections of the data
    * $X\to \tilde{X}$ 
        * This can provide improvement when $n$ is small
        * Decreasing load on the system (high-frequency systems)
        * The maximum variance (variability?) in the model preserved



## Clustering

* Data can be composed of several clusters

* PCA decomposition allows us to recognise such clusters 
    * The principal components will be easy to cluster (linearly separable features)


* PCA works on data centered around zero 
    * Co-centric clusters 


## Outlier Detection

* Extreme version of clustering

The workflow is as follows:

* Perform dimensionality reduction of large-dimensional data set and then find outliers in the low-dimensional space

* This is a tool to decrease the dimensionality of the problem, not a methor per se
    * Turning it into a statistical method requires some further assumption (outlier = extreme values etc)
    * Example: fradulent transactions


***


```python
# Load data, we will use the Boston dataset

import numpy as np
import pandas as pd
from sklearn.datasets import load_boston

dataBoston = load_boston()

# We do not need target variable
df = pd.DataFrame(dataBoston.data, columns=dataBoston.feature_names)
```

```python
# Estimate the PCA, only the first component is needed

# Load PCA
from sklearn.decomposition import PCA

# Data
X = df.values

# PCA with one component
pca = PCA(n_components=1)

# Fit PCA
pca.fit(X)

# PCA component -- first eigenvector * data
pcaX = np.matmul(X,pca.components_[0])
```

```python
# Visualise the outcome
import matplotlib.pyplot as plt
plt.plot(pcaX);
```

```python
# Histogram is more suitable
plt.hist(pcaX, bins=30);
```

```python
Xcluster1=X[pcaX>450]
Xcluster2=X[pcaX<=450]
```

# Kernel PCA

* Let us apply the PCA on the following dataset.



```python
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split

X, y = make_circles(n_samples=1_000, factor=0.3, noise=0.05, random_state=0)
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

ax.scatter(X[:, 0], X[:, 1], c=y)
ax.set_ylabel("x #1")
ax.set_xlabel("x #0")
```

***
* Data are centered around the origin

* Apply PCA
***

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

# fit and project at the same time
X_pca = pca.fit(X).transform(X)
```

```python
fig, (ax1,ax2) = plt.subplots(
    ncols=2, figsize=(14, 4)
)

ax1.scatter(X[:, 0], X[:, 1], c=y)
ax1.set_ylabel("x #1")
ax1.set_xlabel("x #0")
ax1.set_title("Data")

ax2.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
ax2.set_ylabel("Principal component #1")
ax2.set_xlabel("Principal component #0")
ax2.set_title("PCA Projection")
```

***
* Projection leaves data nearly identical

* In the case of the Boston data, we have been able to identify clusters of data based on the first principal component. 

    * In this case, we see two clusters, but we are not able to recongise them
    
    * PCA works with data centered around zero 
    
* Let us use rbf kernel (and then say what we actually did)

***

```python
from sklearn.decomposition import KernelPCA

kernel_pca = KernelPCA(n_components=None, kernel="rbf", gamma=10, alpha=0.1)
```

```python
X_pcaKernel = kernel_pca.fit(X).transform(X)
```

```python
fig, (ax1,ax2) = plt.subplots(
    ncols=2, figsize=(14, 4)
)

ax1.scatter(X[:, 0], X[:, 1], c=y)
ax1.set_ylabel("x #1")
ax1.set_xlabel("x #0")
ax1.set_title("Data")

ax2.scatter(X_pcaKernel[:, 0], X_pcaKernel[:, 1], c=y)
ax2.set_ylabel("Principal component #1")
ax2.set_xlabel("Principal component #0")
ax2.set_title("PCA with rbf Kernel Projection")
```

```python
# Fine tune parameters
kernel_pca = KernelPCA(n_components=None, kernel="rbf", gamma=3, alpha=0.1)

X_pcaKernel = kernel_pca.fit(X).transform(X)

fig, (ax1,ax2) = plt.subplots(
    ncols=2, figsize=(14, 4)
)

ax1.scatter(X[:, 0], X[:, 1], c=y)
ax1.set_ylabel("x #1")
ax1.set_xlabel("x #0")
ax1.set_title("Data")

ax2.scatter(X_pcaKernel[:, 0], X_pcaKernel[:, 1], c=y)
ax2.set_ylabel("Principal component #1")
ax2.set_xlabel("Principal component #0")
ax2.set_title("PCA with rbf Kernel Projection")
```

***
* We have been able to separate two clusters!

# Kernel PCA and its Quantum Counterpart

We are ready to introduce the kernel PCA:

* We have used the sample covariance matrix: 

$${Q}=\frac{1}{n-1}\left({X}'\right)^{T}{X}'$$

which stands for:

$${Q}=\frac{1}{n-1}\sum_{i=1}^{n}\left({x_i}\right){x_i}^{T}\,.$$

* We assume $x_i$ is a column of length $p$, the product is not a dot product.

We can perform a kernel trick and project every observation $x$ from the $p$-dimensional feature space into $\tilde{p}$-dimensional space using a transform $f$.

$$\tilde{x} = f(x)\,.$$

***



***
# Kernel PCA and its Quantum Counterpart

We can calculate: 

$$\tilde{Q}=\frac{1}{n-1}\sum_{i=1}^{n}\left({\tilde{x}_i}\right){\tilde{x}_i}^{T}\,.$$

The sample covariance matrix is a $\tilde{p}\times\tilde{p}$ matrix. 

* This will result in the components in the $\tilde{p}$-dimensional space. 

* We are interested in the original feature space

* Formulation above does not involve kernels 

Let us calculate the PCA in the $\tilde{p}$-dimensional space nevertheless.

Objective:

* all egienvalues $\lambda\geq0$
* corresponding egenvectors $V$

such that

$$\lambda V = \tilde{Q} V\,.$$

Given that $V$ lies in the span of $\tilde{x}_1,\dots,\tilde{x}_n$, we can write the problem:

$$\lambda \tilde{x}_k \cdot V = \tilde{x}_k \cdot\tilde{Q} V\,.$$

for every $k=1,\dots,n$, with decomposition:

$$V = \sum_{i=1}^{n} \alpha_i \tilde{x}_i\,. $$

Now, we can introduce kernel $K$:

$$K(x_i,x_j) = K_{i,j} = \tilde{x}_i \cdot \tilde{x}_i\,,$$

and thus

$$n \lambda K = K^2 \vec{\alpha}\,,$$

with $\vec{\alpha}$ being a column vector of individual $\alpha$'s. 

* The calculation further requires normalisation and some other numerical oprations. However, we do not need the explicit form of $f$, we may work with kernel only. 

* Further, we may use $\tilde{p}=p$, the kernel just warps the space.

```python
# dimension of problem (number of qubits)
nQubits = X_train.shape[1]
# number of repetitions
nReps = 1

# Define Feature map, Pauli-Z evolution circuit 
from qiskit.circuit.library import ZZFeatureMap

featureMap = ZZFeatureMap(feature_dimension=nQubits, reps=nReps, entanglement="linear")
```

```python
from qiskit.utils import QuantumInstance

backendSim = QuantumInstance(
    BasicAer.get_backend("qasm_simulator"), shots=1024
)
```

```python
# define Quantum kernel using the above feature map
kernelQuantum = QuantumKernel(feature_map=featureMap, quantum_instance=backendSim)
```

```python
# Fit (=find eigensystem) and then transform the data using PCA with precalculated kernel

transformed_x = kernelPCAquantum.fit_transform(K)
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

ax.scatter(transformed_x[:, 0], transformed_x[:, 1], c=y_train)
ax.set_ylabel("x #1")
ax.set_xlabel("x #0")
ax2.set_title("PCA Projection with Quantum Kernel, in-sample")
```
